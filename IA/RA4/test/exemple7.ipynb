```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.utils.multiclass import unique_labels
from sklearn.metrics import confusion_matrix
# y_true: valores correctos
# y_pred: valores predichos
# normalize: los mismos valores que para confusion_matrix
# ('true': valor correcto, 'pred': valor predicho, 'all': todos los valores, None: sin normalizar)
# title: titulo
# cmap: color map
def plot_confusion_matrix(y_true, y_pred, normalize=None,
                          title=None, cmap=plt.cm.Blues, rotation=0):
    # Imprimir en forma de texto matriz de confusion y metricas
    if not title:
        if normalize == "true":
            title = "Matriz de confusion (normalizada por valor correcto)"
        elif normalize == "pred":
            title = "Matriz de confusion (normalizada por valor predicho)"
        elif normalize == "all":
            title = "Matriz de confusion (normalizada por totales)"
        else:
            title = "Matriz de confusion (sin normalizar)"
    # Calcular matriz de confusion
    cm = confusion_matrix(y_true, y_pred, normalize=normalize)
    # Utilizar solo las etiquetas que aparecen en los datos
    classes = unique_labels(y_true, y_pred)
    fig, ax = plt.subplots()
    im = ax.imshow(cm, interpolation="nearest", cmap=cmap)
    ax.figure.colorbar(im, ax=ax)
    # Mostrar todos los ticks y etiquetarlos con los entradas de la lista
    ax.set(
        xticks=np.arange(cm.shape[1]),
        yticks=np.arange(cm.shape[0]),
        xticklabels=classes,
        yticklabels=classes,
        title=title,
        ylabel="Valor correcto",
        xlabel="Valor predicho",
    )
    # Rotar las etiquetas para ticks y establecer su alineacion
    plt.setp(ax.get_xticklabels(), rotation=rotation, ha="right", rotation_mode="anchor")
    # Iterar sobre los datos de la matriz para crear anotaciones de texto
    fmt = ".2f" if normalize else "d"
    thresh = cm.max() / 2.0 if cm.size else 0
    for i in range(cm.shape[0]):
        for j in range(cm.shape[1]):
            ax.text(
                j,
                i,
                format(cm[i, j], fmt),
                ha="center",
                va="center",
                color="white" if cm[i, j] > thresh else "black",
            )
    fig.tight_layout()
    return ax
```
